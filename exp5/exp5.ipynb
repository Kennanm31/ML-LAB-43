```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score, roc_auc_score, confusion_matrix, classification_report
```

```python
data=load_breast_cancer()
X=data.data
Y=data.target
print("Classes:",data.target_names)
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
lr=LogisticRegression(max_iter=5000)
lr.fit(X_train,Y_train)
pred_lr=lr.predict(X_test)
print("Logistic TestAccuracy:",accuracy_score(Y_test,pred_lr))
pred_lr_train=lr.predict(X_train)
print("Logistic TrainAccuracy:",accuracy_score(Y_train,pred_lr_train))
precision_lr = precision_score(Y_test, pred_lr)
print("Precision Score:",precision_lr)
recall_lr = recall_score(Y_test, pred_lr)
print("Recall Score:",recall_lr)
```

```python
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
pred_dt_test=dt.predict(X_test)
print("Decision Tree TestAccuracy:",accuracy_score(Y_test,pred_dt_test))
pred_dt_train=dt.predict(X_train)
print("Decision Tree TrainAccuracy:",accuracy_score(Y_train,pred_dt_train))
precision_dt = precision_score(Y_test, pred_dt_test)
print("Decision Tree Precision Score:", precision_dt)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
pred_knn = knn.predict(X_test)
print("KNN Test Accuracy:", accuracy_score(Y_test, pred_knn))
pred_knn_train = knn.predict(X_train)
print("KNN Train Accuracy:", accuracy_score(Y_train, pred_knn_train))
precision_knn = precision_score(Y_test, pred_knn)
print("KNN Precision Score:", precision_knn)
```

```python
nb = MLPClassifier(max_iter=5000)
nb.fit(X_train, Y_train)
pred_nb = nb.predict(X_test)
print("MLPClassifier TestAccuracy:", accuracy_score(Y_test, pred_nb))
pred_nb_train = nb.predict(X_train)
print("MLPClassifier TrainAccuracy:", accuracy_score(Y_train, pred_nb_train))
precision_nb = precision_score(Y_test, pred_nb)
print("MLPClassifier Precision Score:", precision_nb)
```

```python
nn = MLPClassifier(max_iter=5000)
nn.fit(X_train, Y_train)
pred_nn = nn.predict(X_test)
print("Neural Network Test Accuracy:", accuracy_score(Y_test, pred_nn))
pred_nn_train = nn.predict(X_train)
print("Neural Network Train Accuracy:", accuracy_score(Y_train, pred_nn_train))
precision_nn = precision_score(Y_test, pred_nn)
print("Neural Network Precision Score:", precision_nn)
```

```python
precision_lr = precision_score(Y_test, pred_lr)
```

```python
cm_lr = confusion_matrix(Y_test, pred_lr)
print("Logistic Regression Confusion Matrix:\n", cm_lr)
TN_lr, FP_lr, FN_lr, TP_lr = cm_lr.ravel()
type1_error_lr = FP_lr / (FP_lr + TN_lr)
type2_error_lr = FN_lr / (TP_lr + FN_lr)
print("Logistic Regression Type 1 Error:", type1_error_lr)
print("Logistic Regression Type 2 Error:", type2_error_lr)
print("Type-I error rate in Logistic Regression in %:", type1_error_lr*100)
print("Type-II error rate in Logistic Regression in %:", type2_error_lr*100)
```

```python
cm_dt = confusion_matrix(Y_test, pred_dt_test)
print("Decision Tree Confusion Matrix:\n", cm_dt)
TN_dt, FP_dt, FN_dt, TP_dt = cm_dt.ravel()
type1_error_dt = FP_dt / (FP_dt + TN_dt)
type2_error_dt = FN_dt / (TP_dt + FN_dt)
print("Decision Tree Type 1 Error:", type1_error_dt)
print("Decision Tree Type 2 Error:", type2_error_dt)
print("Type-I error rate in Decision Tree in %:", type1_error_dt*100)
print("Type-II error rate in Decision Tree in %:", type2_error_dt*100)
```

```python
cm_knn = confusion_matrix(Y_test, pred_knn)
print("KNN Confusion Matrix:\n", cm_knn)
TN_knn, FP_knn, FN_knn, TP_knn = cm_knn.ravel()
type1_error_knn = FP_knn / (FP_knn + TN_knn)
type2_error_knn = FN_knn / (TP_knn + FN_knn)
print("KNN Type 1 Error:", type1_error_knn)
print("KNN Type 2 Error:", type2_error_knn)
print("Type-I error rate in KNN in %:", type1_error_knn*100)
print("Type-II error rate in KNN in %:", type2_error_knn*100)
```
